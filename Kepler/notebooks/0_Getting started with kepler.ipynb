# Import

```python
import pandas as pd
from keplergl import KeplerGl
import geopandas as gpd
```

# 1. New York PD complaints data

# Load data

```python
df = gpd.read_file("../data/NYPD_Complaint_Data_Current__Year_To_Date_.csv")
df.head()
```

```python
df['CMPLNT_FR_DT'] = pd.to_datetime(df['CMPLNT_FR_DT'],errors='coerce')
```

```python
# Cut down to 2018
df_18 = df[df['CMPLNT_FR_DT'].dt.year ==2018]
```

```python
df_18
```

# Export for later use

```python
df_18.to_csv('../data/NYPD_COMPLAINTS_2018.csv')
```

# Getting started
## Create a basemap

```python
map = KeplerGl(height=600,width=800)
map
```

## Add data
* the easiest way to do so is to convert it to a geodataframe which we can do with geopandas

```python
# use 2018 data
df = df_18
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))
map.add_data(data=gdf, name= "crimes")
```
